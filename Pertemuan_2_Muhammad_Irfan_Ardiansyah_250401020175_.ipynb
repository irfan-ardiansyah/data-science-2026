```python
import pandas as pd
import numpy as np

URL = "http://raw.githubusercontent.com/datasciencedojo/datasets/refs/heads/master/titanic.csv"

df = pd.read_csv(URL)
```

```python
print(df.shape, df.columns.tolist())
print("-----------------------------")
print(df.isnull().sum())
```

```python
# Filtering — Penumpang Wanita Kelas 1
wk1 = df[(df['Sex']=='female') & (df['Pclass']==1)]
wk1

# Rounding result
```

```python
mean_survived = wk1['Survived'].mean()
mean_age = wk1['Age'].mean()

print("============== Raw Result ===============")
print("mean of survived:", mean_survived)
print("mean of age:", mean_age)

print("\n============== Rounded Result ===============")
print("mean of survived:", round(mean_survived, 3))
print("mean of age:", round(mean_age, 3))
```

```python
# Groupby — Tingkat Keselamatan per Kelas
hasil = df.groupby('Pclass').agg(total=('PassengerId','count'), selamat=('Survived','mean')).round(3)
print(hasil)
```

```python
# Jawaban
print("1. Jumlah baris & kolom")
print("baris:", df.shape[0])
print("kolom", df.shape[1])
print("\n")

print("2. Kolom yang bernilai kosong")
print("Kolom kosong:", df.columns[df.isnull().sum() != 0].tolist())
print("\n")

print("3. Persentase penumpang selamat")
survived = len(df[(df["Survived"]==1)]) / df.shape[0] * 100
print(round(survived, 2), "%")
print("\n")


print("4. Rata-rata usia penumpang wanita kelas 1 yang selamat")
print(round(mean_age, 2), "%")
print("\n")


print("5. Kelas penumpang mana yang tingkat keselamatan tertinggi")
print("Kelas ", hasil["selamat"].idxmax() )
print("\n")


print("6. Tampilkan hasil groupby('Pclass')['Survived'].mean()")
print(df.groupby("Pclass")['Survived'].mean().round(2) )
print("\n")
```
